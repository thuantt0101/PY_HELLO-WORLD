```python
import numpy as np
import matplotlib.pyplot as plt

# Tinh phan phoi chuan Normal Distribution
x = np.random.normal(5.0, 1.0, 100000)

# Ve bieu do phan phoi chuan
plt.hist(x, 1000)
plt.show()
```

# Giải thích biểu đồ
Quan sát biểu đồ có hình dạng giống hình chuông ( bell ), do đó được gọi là 
đường cong hình chuông ( bell curves )

Biểu đồ trên thể hiện phân phối dữ liệu của 100000 giá trị với giá trị trung bình là 5.0 
& độ lệch chuẩn là 1.0

Biểu đồ trên được vẽ bởi 1000 bars ( 1000 là thanh hoặc cột )

Nhìn vào biểu đồ trên chúng ta thấy các giá trị hầu hết chủ yếu nằm trong khoảng từ 4.0 đến 6.0 
với giá trị xấp xỉ gần gằng 5.0
